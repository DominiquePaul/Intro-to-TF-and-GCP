# Building an Image Classifier with Tensorflow
In this script we will create an image classifier on a dataset of cats and dogs with tensorflow using a model commonly termed a convolutional neural network

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
# from tensorflow.keras.layers import Conv2D, MaxPooling2D
from sklearn.model_selection import train_test_split
import os
import cv2
from tqdm import tqdm
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pickle
```

### Defining the size of our images
We choose our images to have the size 64x64

```python
IMG_SIZE = 64
```

## 1. Preparing the Data

### We define some function to load the images easier

```python
def load_image(img_path):
    """This function loads in the data applies some transformations but skips the file
    should it be corrupted or any other error should occur
    """
    try:
        img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE) # convert the image from RGB to Greyscale
        img = cv2.resize(img,(IMG_SIZE,IMG_SIZE)) # resize the image to a 64x64 pixel size
        return(img)
    except: # if the code in the try statement throws an error then the following code block is executed.
        print(img_path)
        return(None)

def create_training_set(path, label):
    """This function traverses through a folder an gets all images which it appends to 
    an array. It also creates a second array with the same length containing the labels.
    """
    images_ = [] # define an empty list, we will populate this later with the image values
    for subdir, dirs, files in os.walk(path): # this returns all subfolders and files of the files in the specified path
        for i,file in enumerate(tqdm(files)):
            image_filepath = os.path.join(os.getcwd(),subdir,file) # reconstruct the specific file path 
            img = load_image(image_filepath) # we use the previously defined function to load the individuals images
            if img is not None: # we only append the file to our image list if the value is not none
                images_.append(img)
    # we create a new list that has the same length as the list images_ but only contains 
    # the value of the variable label
    labels_ = [label] * len(images_) 
    return(np.array(images_), np.array(labels_))
```

### The folder paths with our images

```python
path1 = "Data/Cats"
path2 = "Data/Dogs"
```

### Lets have a look what one of the files look like as an original
We see that the original image has a height of 329, a width of 500 and 3 colour channels (Red, Green, Blue)

```python
example_file = "../Data/sample_image2.jpg"
img = cv2.imread(example_file)
print(img.shape)
# We need the cv2.cvtColor command because CV2 saves files in Blue Green Red (the reverse order)
# and wee need to convert the file to RGB before we display it
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

### We load both the cat and dog images and assign them the labels 0 and 1 respectively

```python
imgs_cats, labels_cats = create_training_set(path1,0)
imgs_dogs, labels_dogs = create_training_set(path2,1)
```

### Creating our training and test data
We merge the images and labels together respectively to then shuffle them and create a training and test set

```python
imgs = np.concatenate((imgs_cats, imgs_dogs))
labels = np.concatenate((labels_cats, labels_dogs))

x_train, x_test, y_train, y_test = train_test_split(imgs,
                                                    labels,
                                                    train_size=0.8,
                                                    test_size=0.2)
```


### Reshaping and Normalizing
We reshape our data

```python
print(x_train.shape)

x_train = np.array(x_train).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
x_test = np.array(x_test).reshape(-1, IMG_SIZE, IMG_SIZE, 1)

print(x_train.shape)
```

### Normalizing the data 
We normalize the data to avoid the model being less sensitive to the scale of the features

```python
x_train = tf.keras.utils.normalize(x_train, axis=1)  # scales data between 0 and 1
x_test = tf.keras.utils.normalize(x_test, axis=1)  # scales data between 0 and 1
```

## Defining our Convolutional Neural Network
We are using a convolutional neural network which is easy to assemble with Keras, a high-level programming interface for tensorflow hiding many operations and reducing the code to the most basic funtions

A convolutional neural network uses a mix of convolutional layers (layers which scan small parts of the pixels for specific patterns), activation functions (which add non-linearity to our model, i.e. enabling it learn even more complicated patterns) and pooling layers (which reduce the size of the layers by only conserving the greatest values).

**More Information**  
A lengthy and very good explanation can be found here: http://cs231n.github.io/convolutional-networks/#pool  
A nice video on CNNs: https://www.youtube.com/watch?v=FmpDIaiMIeA&t  
A great visualisation: http://scs.ryerson.ca/~aharley/vis/conv/flat.html  

```python
def create_model():
    model = Sequential()

    model.add(Conv2D(256, (3, 3), input_shape=(IMG_SIZE, IMG_SIZE, 1)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(256, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(128, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors
    model.add(Dense(64))

    model.add(Activation('relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1))
    model.add(Activation('sigmoid'))

    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])

    return(model)
```

### Training the Model

```python
model = create_model() # we instantiate the model
model.fit(x_train, # the input values to train on
          y_train, # the output values to train on 
          batch_size=32, # the amount of images we want to feed at the same time
          epochs=20, # the amount of epochs we want to train out network
          validation_split=0.3)
```

### Saving weights and having a look under the hub
By saving the weights we can load the model again anytime later and dont have to train it again. This saves us a lot of time. Especially when we consider that some models can even take several days to train!

```python
model.save_weights("Model-Checkpoints/model-checkpoints2")
model.summary() # shows us the amount of parameters we are training at each layer
```

### Creating a new version and loading the pre-trained weights

```python
new_model = create_model()
new_model.load_weights("Model-Checkpoints/model-checkpoints2")
```

### Making new predictions
With our new model we can now easily make new predictions for individual new images. Either the individual score or the class

```python
img = load_image("/Users/dominiquepaul/xJob/4-Lecture/Data/Cats/5.jpg")
print(new_model.predict(img))
print(new_model.predict_classes(img))
```

### Confusion Matrix
Lets have a look at how the model makes the mistakes by looking at the confusion matrix

```python
# import a ready-to-use function from scikit learn
import sklearn.metrics as metrics 
```

```python
preds_test = new_model.predict_classes(x_test)
confusion_matrix = metrics.confusion_matrix(y_true=y_test, y_pred=preds_test)
print(confusion_matrix)
```

#### Downloading and formatting an image from the web
We write a function that 
* loads an image from an URL 
* converts its color to grey 
* resizes it
* normalizes it
* and reshapes it to the correct input siez

```python
def load_from_url(link):
    img = io.imread(link)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img,(IMG_SIZE,IMG_SIZE))
    img = tf.keras.utils.normalize(img, axis=1)
    img = img.reshape(-1, IMG_SIZE, IMG_SIZE, 1)
    return(img)
```

We load the image and have a peak at it

```python
img_url2 = "https://cdn8.bigcommerce.com/s-rj0z9yqukq/images/stencil/original/uploaded_images/sayslove.jpg?t=1518480203"
image = load_from_url(img_url2)  # call our custom function
# displaying the image, we need to change the color-map for matplotlib to know that it is a grey image
plt.imshow(image, cmap = 'gray') 
```
