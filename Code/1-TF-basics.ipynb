# 1. Tensorflow Basics

### Importing tensorflow

```python
import tensorflow as tf

import cv2 # package used for working with images
import matplotlib.pyplot as plt 
```

### Data is stored in tensors. We can create tensors like this:

```python
a = tf.constant([1],name='Const1')
b = tf.constant([2], name="Const2")
```

### We can add them together like this

```python
c = tf.add(a,b)
```

### But printing the result gives us an unexpected result:

```python
print(c)
```

This is because TF is **lazily evaluated**. This is because so far we have only created the graph of our application. We have to explicitly tell TF to evaluate the values. 

Tensorflow represents numeric computations as a graph. **EVERY computation is Tensorflow will be represented as a graph**. Nodes are operations. Edges are tensors (i.e. data) flowing between the graphs. Each operation is a function that can be evaluated at that point. We start off by using different values (constants, variables, placeholders) to construct a graph. But this is just the theoretical construct. No data flows until we run the graph in a session.

![title](https://cdn-images-1.medium.com/max/1600/1*G9MXGOM2jWl3SOXuqRlV3A.png)

### Running the Graph in a Tensorflow Session
We can deploy our code in a so-called session. A session is a binding to a computation instance that supports the operations of our computations in our graph, such as a CPU or GPU - in our case a CPU.

```python
with tf.Session() as sess:
    print(sess.run(c))
```
